# Model Building

## 2.1 Define constatns

```python
N_CLASSES = 26
N_EPOCHS = 10
INPUT_SHAPE = (784, )
LEARNING_RATE = 0.01
BATCH_SIZE = 32
```

## 2.2 Define measurement functionalities 

### 2.2.1 Recall Function

The recall function is a way to measure the performance of a model. It is a measure of how many of the positive examples were correctly identified.

```python
def recall(y_exac, y_pred):
    exac_pos = keras.backend.sum(keras.backend.round(keras.backend.clip(y_exac * y_pred, 0, 1)))
    poss_pos = keras.backend.sum(keras.backend.round(keras.backend.clip(y_exac, 0, 1)))
    return exac_pos / (poss_pos + keras.backend.epsilon())
```

### 2.2.2 Precision Function

The precision function is a way to measure the performance of a model. It is a measure of how many of the positive examples were correctly identified.

```python
def precision(y_exac, y_pred):
    exac_pos = keras.backend.sum(keras.backend.round(keras.backend.clip(y_exac * y_pred, 0, 1)))
    pred_pos = keras.backend.sum(keras.backend.round(keras.backend.clip(y_pred, 0, 1)))
    return exac_pos / (pred_pos + keras.backend.epsilon())
```

### 2.2.3 F1 Score Function

The F1 score is a measure of a model's performance. It is the harmonic mean of the precision and recall.

```python
def f1_score(y_exac, y_pred):
    precision_val = precision(y_exac, y_pred)
    recall_val = recall(y_exac, y_pred)
    return 2 * (precision_val * recall_val) / (precision_val + recall_val + keras.backend.epsilon())
```

## 2.3 Create Network

```python
results = {}
```

```python
model = Sequential()
model.add(Dense(units=256, activation='relu', input_shape=INPUT_SHAPE, kernel_initializer='he_normal'))
model.add(Dense(units=128, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=64, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=26, activation='softmax', kernel_initializer='he_normal'))
```

```python
model.compile(optimizer=SGD(learning_rate=LEARNING_RATE), loss='categorical_crossentropy', metrics=['accuracy', recall, precision, f1_score])
```

```python
model.summary()
```

### 2.3.1 Fit Model

```python
results['gcd'] = model.fit(X_train, y_train, epochs=N_EPOCHS, batch_size=BATCH_SIZE)
```

## 4.2 Effect of Optimizer

### 4.2.1 What is SGD?
**Stochastic Gradient Descent** is an iterative method for optimizing an objective function. It minimizes the loss function by updating the parameters of a model using gradient descent.
Both *statistical estimation* and *machine learning* consider the problem of *minimizing* and *objective function* that has the form:

$$Q(w) = \min_{w} \frac{1}{N} \sum_{i=1}^N \mathcal{L}(y_i, f(x_i, w))$$


### 4.2.2 What is Momentum?
SGD with momentum is method which helps accelerate gradients vectors in the right directions, thus leading to faster converging. It is one of the most popular optimization algorithms and many state-of-the-art models are trained using it. What we want to do with this data is, instead of using it, we want some kind of ‘moving’ average which would ‘denoise’ the data and bring it closer to the original function. The form of the moving average is:
$$w_{t+1} = w_t + \alpha \Delta w_t$$

## 4.3 Effect of Epochs

As the number of epochs increases, more number of times the weight are changed in the neural network and the curve goes from underfitting to optimal to overfitting curve.

One of the critical issues while training a neural network on the sample data is Overfitting. When the number of epochs used to train a neural network model is more than necessary, the training model learns patterns that are specific to sample data to a great extent. This makes the model incapable to perform well on a new dataset. This model gives high accuracy on the training set (sample data) but fails to achieve good accuracy on the test set. In other words, the model loses generalization capacity by overfitting to the training data.

To mitigate overfitting and to increase the generalization capacity of the neural network, the model should be trained for an optimal number of epochs. A part of training data is dedicated for validation of the model, to check the performance of the model after each epoch of training. Loss and accuracy on the training set as well as on validation set are monitored to look over the epoch number after which the model starts overfitting.

## 4.4 Effect of Loss Function

### 4.4.1 MSE Loss Function

The mean squared error loss function is used to measure the performance of a model. It is a measure of how far the model is from the ideal.

Mean squared error (MSE) loss is calculated by taking the difference between `y` and our prediction, then square those values. We take these new numbers (square them), add all of that together to get a final value, finally divide this number by y again. This will be our final result.

The formula for calculating mean squared error loss is as follows:
$$\frac{1}{N} \sum_{i=1}^N (y_i - f(x_i))^2$$

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['accuracy', recall, precision, f1_score])
```

```python
results['mse'] = model.fit(X_train, y_train, epochs=N_EPOCHS, batch_size=BATCH_SIZE)
```

### 4.4.2 Why the MSE function is not good for classification?

There are two reasons why Mean Squared Error(MSE) is a bad choice for binary classification problems:

First, using MSE means that we assume that the underlying data has been generated from a normal distribution (a bell-shaped curve). In Bayesian terms this means we assume a Gaussian prior. While in reality, a dataset that can be classified into two categories (i.e binary) is not from a normal distribution but a Bernoulli distribution (Check out the answer to “Where did the Binary Cross-Entropy Loss Function come from?” for more details and also for an intro to maximum likelihood estimation).

Secondly, the MSE function is non-convex for binary classification. In simple terms, if a binary classification model is trained with MSE Cost function, it is not guaranteed to minimize the Cost function. This is because MSE function expects real-valued inputs in range(-∞, ∞), while binary classification models output probabilities in range(0,1) through the sigmoid/logistic function.

## 4.5 Effect of Regularization

Regularization refers to techniques that are used to calibrate machine learning models in order to minimize the adjusted loss function and prevent overfitting or underfitting.

A regression model that uses L1 regularization technique is called Lasso Regression and model which uses L2 is called Ridge Regression.

### 4.5.1 L1 Regularization

Lasso Regression (The Least Absolute Shrinkage and Selection Operator) adds “absolute value of magnitude” of coefficient as penalty term to the loss function.
$$\frac{1}{N} \sum_{i=1}^N \|w_i\|$$

### 4.5.2 L2 Regularization

Ridge regression adds “squared magnitude” of coefficient as penalty term to the loss function. Here the highlighted part represents L2 regularization element.
$$\frac{1}{N} \sum_{i=1}^N (y_i - f(x_i))^2 + \lambda \sum_{j=1}^M |w_j|^2$$

```python
model = Sequential()
model.add(Dense(units=256, activation='relu', input_shape=INPUT_SHAPE, kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.0001)))
model.add(Dense(units=128, activation='relu', kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.0001)))
model.add(Dense(units=64, activation='relu', kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.0001)))
model.add(Dense(units=26, activation='softmax', kernel_initializer='he_normal'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy', recall, precision, f1_score])
```

```python
results['regl2'] = model.fit(X_train, y_train, epochs=N_EPOCHS, batch_size=BATCH_SIZE)
```

```python
pd.DataFrame(results['regl2'].history).to_csv('regl2.csv')
```

### 4.5.3 Dropout

Dropout is a technique that is used to prevent overfitting. It is a technique that randomly sets a fraction of the input units to zero at each training step. This prevents a model from overfitting to the training data.

When you increase dropout beyond a certain threshold, it results in the model not being able to fit properly. Intuitively, a higher dropout rate would result in a higher variance to some of the layers, which also degrades training. Dropout is like all other forms of regularization in that it reduces model capacity.

```python
model = Sequential()
model.add(Dense(units=256, activation='relu', input_shape=INPUT_SHAPE, kernel_initializer='he_normal'))
model.add(Dropout(0.1))
model.add(Dense(units=128, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.1))
model.add(Dense(units=64, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.1))
model.add(Dense(units=26, activation='softmax', kernel_initializer='he_normal'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy', recall, precision, f1_score])
```

```python
results['dropout'] = model.fit(X_train, y_train, epochs=N_EPOCHS, batch_size=BATCH_SIZE)
```

```python
pd.DataFrame(results['dropout'].history).to_csv('dropout.csv')
```

```python
plt.plot(results['dropout'].history['loss'])
plt.plot(results['dropout'].history['accuracy'])
plt.plot(results['dropout'].history['recall'])
plt.plot(results['dropout'].history['precision'])
plt.plot(results['dropout'].history['f1_score'])
plt.legend(['loss', 'accuracy', 'recall', 'precision', 'f1_score'])
plt.show()
```
